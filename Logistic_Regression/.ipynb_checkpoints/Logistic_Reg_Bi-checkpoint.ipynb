### 4. The variables we'll use to build logistic regression model are :
        1. Satisfaction Level 
        2. Average Month Hours
        3. Promotion Last 5 years
        4. Salary

```python
new_df = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]

new_df
```

#### As the salaries coloums are string we will modify out df to make use of salary.

```python
sal_dummies = pd.get_dummies(new_df.salary, prefix = "salary")
```

```python
new_df = pd.concat([new_df, sal_dummies], axis='columns')
new_df.drop('salary',axis = 'columns',inplace = True)

new_df
```

```python
X = new_df

X
```

```python
y = df.left
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

### 5. Measuring the accuracy of model

```python
model.score(X_test,y_test)
```

#### So accuracy of the model is around 77.97%
